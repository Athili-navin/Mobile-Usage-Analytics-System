## 1. Setup & Data Loading

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 5)

df = pd.read_csv("datamobile_usage.csv")
df.head()
```

## 2. Data Preprocessing

Check structure, types, missing values and duplicates.

```python
df.info()
```

```python
print("Missing values per column:")
print(df.isnull().sum())
print("\nDuplicate rows:", df.duplicated().sum())
```

```python
# Clean column names and drop any duplicates / nulls (defensive — dataset is already clean)
df.columns = [c.strip() for c in df.columns]
df = df.drop_duplicates().dropna()
df.describe().round(2)
```

## 3. Feature Engineering

Derive a few extra analytical columns used later in the dashboard.

```python
df["Age Group"] = pd.cut(df["Age"], bins=[0,25,35,45,55,200],
                         labels=["18-25","26-35","36-45","46-55","56+"])
df["App Usage Share (%)"] = (df["App Usage Time (min/day)"] /
                              (df["Screen On Time (hours/day)"] * 60) * 100).clip(upper=100).round(1)
df["Data per App (MB/day)"] = (df["Data Usage (MB/day)"] / df["Number of Apps Installed"]).round(2)
df.head()
```

## 4. Univariate Analysis — Distributions

```python
fig, axes = plt.subplots(2, 2, figsize=(13, 9))
sns.histplot(df["Screen On Time (hours/day)"], kde=True, ax=axes[0,0], color="#6C5CE7")
axes[0,0].set_title("Screen On Time Distribution")
sns.histplot(df["App Usage Time (min/day)"], kde=True, ax=axes[0,1], color="#0984E3")
axes[0,1].set_title("App Usage Time Distribution")
sns.histplot(df["Data Usage (MB/day)"], kde=True, ax=axes[1,0], color="#E17055")
axes[1,0].set_title("Data Usage Distribution")
sns.histplot(df["Age"], kde=True, ax=axes[1,1], color="#00B894")
axes[1,1].set_title("Age Distribution")
plt.tight_layout()
plt.show()
```

## 5. Categorical Breakdown

```python
fig, axes = plt.subplots(1, 3, figsize=(16, 4))
df["Device Model"].value_counts().plot(kind="bar", ax=axes[0], color="#6C5CE7")
axes[0].set_title("Users by Device Model")
df["Operating System"].value_counts().plot(kind="bar", ax=axes[1], color="#0984E3")
axes[1].set_title("Users by Operating System")
df["Gender"].value_counts().plot(kind="bar", ax=axes[2], color="#E17055")
axes[2].set_title("Users by Gender")
plt.tight_layout()
plt.show()
```

## 6. Screen Time Analytics

```python
fig, ax = plt.subplots(figsize=(9, 5))
df.groupby("Device Model")["Screen On Time (hours/day)"].mean().sort_values().plot(kind="barh", ax=ax, color="#6C5CE7")
ax.set_title("Average Screen Time by Device")
ax.set_xlabel("Hours/day")
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(9,5))
sns.boxplot(data=df, x="User Behavior Class", y="Screen On Time (hours/day)", hue="User Behavior Class", legend=False, ax=ax, palette="viridis")
ax.set_title("Screen Time by Behavior Class")
plt.tight_layout()
plt.show()
```

## 7. App Usage Analysis

```python
fig, ax = plt.subplots(figsize=(9,5))
sns.scatterplot(data=df, x="Number of Apps Installed", y="App Usage Time (min/day)",
                 hue="User Behavior Class", palette="viridis", ax=ax)
ax.set_title("Apps Installed vs Daily App Usage Time")
plt.tight_layout()
plt.show()
```

## 8. Internet & Data Consumption

```python
fig, ax = plt.subplots(figsize=(9,5))
df.groupby("Device Model")["Data per App (MB/day)"].mean().sort_values().plot(kind="barh", ax=ax, color="#E17055")
ax.set_title("Average Data Consumption per Installed App, by Device")
ax.set_xlabel("MB/day per app")
plt.tight_layout()
plt.show()
```

## 9. Correlation & Behavior Insights

```python
numeric_cols = ["App Usage Time (min/day)", "Screen On Time (hours/day)",
                 "Battery Drain (mAh/day)", "Number of Apps Installed",
                 "Data Usage (MB/day)", "Age", "User Behavior Class"]
fig, ax = plt.subplots(figsize=(8,6))
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap="coolwarm", center=0, ax=ax)
ax.set_title("Correlation Heatmap — Usage Metrics")
plt.tight_layout()
plt.show()
```
